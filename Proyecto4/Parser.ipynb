```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv("adultIncome.csv")
df.head(10)
```

```python
df["education"].unique()
```

```python
ndf = pd.DataFrame()
#ndf["age"] = df["age"].tolist()
ndf["wc"] = df["wc"]
ndf["education"] = df["education"]
ndf["marital status"] = df["marital status"]
ndf["race"] = df["race"]
ndf["gender"] = df["gender"]
#ndf["hours per week"] = df["hours per week"].tolist()
ndf["IncomeClass"] = df["IncomeClass"]
ndf.head(10)
```

```python
ndf = ndf.apply( LabelEncoder().fit_transform )
ndf
```

```python
df_parser = pd.DataFrame()
df_parser["age"] = df["age"].tolist()
df_parser["wc"] = ndf["wc"]
df_parser["education"] = ndf["education"]
df_parser["marital status"] = ndf["marital status"]
df_parser["race"] = ndf["race"]
df_parser["gender"] = ndf["gender"]
df_parser["hours per week"] = df["hours per week"].tolist()
df_parser["IncomeClass"] = ndf["IncomeClass"]
df_parser.head(10)
```

```python
df_parser.to_csv("dataframe.csv", index=False)
```

```python
cant_train = int(0.75*df_parser.shape[0])
train = df_parser.iloc[:cant_train]
test = df_parser.iloc[cant_train:]
test.index = pd.RangeIndex(len(test.index))
```

```python
train.to_csv("train.csv", index=False)
test.to_csv("test.csv", index=False)
```
